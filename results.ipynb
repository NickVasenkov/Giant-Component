# Avoiding a giant component

## Introduction

### The game description

Choose integers n, m, s.t. m<n.

Start with an empty graph of a size n.

In each step:

    Choose two pairs of unconnected nodes at random.
    Choose which pair to connect with an edge.
    
The game lasts until a component of size m emerges.

The goal is to maximize the number of steps (i.e., avoid the emergense of a component of size m).


### Our work

We compared average performances of some human-made algorithms with average performances of AI agents.

### Human-made algorithms

Currently, the following algorithms are implemented:

1) Random: Always chooses a first pair (note, the order of pairs is arbitrary).
2) Sum rule: Chooses either a pair that belongs to the same component, or a pair with the least sum of corresponding components.
3) Product rule: Chooses either a pair that belongs to the same component, or a pair with the least product of corresponding components.

The product rule is considered to be nearly optimal, with a few algorithms that seem to do slighlty better (Beveridge, Andrew, et al. “Product Rule Wins a Competitive Game.” Proceedings of the American Mathematical Society, vol. 135, no. 10, 2007, p. 3062).

### AI Agents

Currently, we are using the RLLib's implementation of the Proximal Policy Optimization algorithm. Please, refer to the '.py' files for the details.
